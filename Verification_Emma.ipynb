```python
def GetSunMoonSetTime(d,lat,lon,mode='next'):
    # specify your date (a time is needed so let's put midnight)
    #d=Time(jd)
    #mode should be ‘next’, ‘previous’, or ‘nearest’
    
    coords=EarthLocation.from_geodetic(lon,lat)
    
    #create an observer object from the location
    obs = Observer(location=coords, timezone="UTC")
    #print("Getting sunset and moonset times for specified date:")
    moonset=obs.moon_set_time(time=d,which=mode)
    sunset=obs.sun_set_time(time=d,which=mode)
    return sunset,moonset

def GetSunMoonRiseTime(jd):
    # specify your date (a time is needed so let's put midnight)
    d=Time(jd,lat,lon,mode='next')
    #mode should be ‘next’, ‘previous’, or ‘nearest’
    
    coords=EarthLocation.from_geodetic(lon,lat)
    
    #create an observer object from the location
    obs = Observer(location=coords, timezone="UTC")
    #print("Getting sunset and moonset times for specified date:")
    moonrise=obs.moon_rise_time(time=d,which=mode)
    sunrise=obs.sun_rise_time(time=d,which=mode)
    return sunrise,moonrise
    
def Tb(Ts,Tm): # calculate best observation time from sunset and moonset
    # note reverse this for morning observations!
    # Ts= sunset time
    # Tm= moonset time
    # used 4/9 assumption
    lag=Tm-Ts
    Tb=Ts+(4./9.)*lag
    return Tb
```
